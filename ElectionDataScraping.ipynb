### Code to generate links to each page  

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re

base = "https://myneta.info/loksabha2014/" # Base Link for the data 1st level that contains links for other pages

req = Request(base)
html_page = urlopen(req)

soup = BeautifulSoup(html_page, "html")

#print(soup)

html_links = soup.find_all('table')[2:]#.find_all('a')#.find_all('tr')
# The page contains three tables, and we need the third one for our data, so we get all the table data since the 3rd table

# links = []
# for link in soup.findAll('a'):
#     links.append(link.get('href'))

# print(links)

# Link Generator 

tr_links = html_links[0].find_all('tr') 
# Find the rows of data defined by the table tag <tr> which will contain both <th> and <td> tags

paths = [] # Empty list to append and contain paths

for tr_link in tqdm(tr_links): 
    # Looping through each of the found list of <tr> tags data discovered
    
    for td in tr_link.find_all('td'):
        # finding each <td> tag in each of <tr> tags
        rest = '' # empty string to store the rest of the link after base
        
        for a in td.find_all('a'): # finding <a> tags in each of <td> tags 
            href_list = a.get('href').split(base) # getting href attribute from <a> tags
            # The split() function is used after base link as many links are already full links and we need to get the rest of
            # link from them.
            
            if len(href_list) == 2: 
                # if the split function has 2 values (0 - empty and 1 - rest of link) i.e.,  it contains full link
                href = href_list[1]
            else:
                # if not then get the first element (index=0) as the string will not be split
                href = href_list[0]
            path = base + href # combine link
            print(path) 
            paths.append(path) # append to paths
            
print(paths)
```
