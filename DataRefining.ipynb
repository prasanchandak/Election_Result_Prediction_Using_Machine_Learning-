**Finding discrepencies**

We have discrepencies in names of candidates. We cannot completely check the names of each candidate and compare and correct it. So, we do the next best thing. In case a winner is missing, we only correct the winner so we can make sure that the model is up to date.

Discrepencies could be in the form of spelling mistakes OR honorific titles such as Mr., Dr., Smt etc. OR they can be of the form where first name and last name sequence is reversed OR the full name is abbreviated.

To correct this we need to have a common value for candidate name in both dataframes before combining them. We can do this by locating the discrepency constituency, getting winner rows from both dataframes and copying one to another.

THERE MUST BE ONE WINNER IN EACH CONSTITUENCY.

If candidate is a winner, "Winner" column = 1. 

We group candidates constituency-wise and then find mean of each. If mean is 0, then the winner is missing from that constituency. 

Empty results for df.loc[] are because of unmatching constituency names (they are used in df.loc[] to locate values) they also need to be changed.

LokDhaba dataset contains values in current times (including Telangana) while MyNeta data is of the old format. (This does not impact the constituency, but it does impact manually replacing them (eg. chelvella).

**Solution:** 

Add this code after the dataset has been cleared. Use the previously combined data with missing winner=1 values as a reference dataframe. It will be useful later. 
Check if the length of both queries used in df.loc[] is same. If same then replace the value in lokdhaba with that of scraped dataset. If length turns out to be 0, then replace all constituency names at that column.
If any length is 0, ignore rows since the values in dataframe are not common.
If length of any one or both is 2, just replace the first values of each.

```python
def combine(a, b):
    
    col = ["Candidate", "Party", "CriminalCases", "Education", "Age", "Gender", "TotalAssets", "Liabilities", "Winner", "State", 
       "Constituency", "Year", "Category", "Votes", "ValidVotes", "VoteSharePercentage", "PartyID"]

    combined = pd.DataFrame(columns=col)

    i = 0
    j = 0
    
    r = 0
    
    if a.shape[0]<b.shape[0]:
        r = b.shape[0]
    
    else:
        r = a.shape[0]
    
    a_unmatched = 0
    b_unmatched = 0
    # Counters for calculating unmatched values in a and b
    
    print("\nShape(a): ", a.shape)
    print("Shape(b): ", b.shape)

    for k in tqdm(range(r)):
        if a["Candidate"][i] == b["Candidate"][j]:
            combined = combined.append({"Candidate":a["Candidate"][i], "Party":a["Party"][i], 
                                        "CriminalCases":a["CriminalCases"][i], "Education":a["Education"][i], "Age":a["Age"][i],
                                        "Gender": b["Sex"][j], "TotalAssets":a["TotalAssets"][i], "Liabilities":a["Liabilities"][i], 
                                        "Winner":a["Winner"][i], "State":a["State"][i], "Constituency":a["Constituency"][i], 
                                        "Year":a["Year"][i], "Category":b["Candidate_Type"][j], "Votes":b["Votes"][j], 
                                        "ValidVotes":b["Valid_Votes"][j], "VoteSharePercentage":b["Vote_Share_Percentage"][j],
                                        "PartyID":b["Party_ID"][j]}, ignore_index=True)
            # Append value to the combined dataframe as a dictionary value if the names match and move on to next value            
            i += 1
            j += 1
            
        elif a["Candidate"][i] < b["Candidate"][j]:
            i += 1
            a_unmatched += 1
            # If alphabetical value of a is less than b then it means that the a name is not in b as we have an alphabetically 
            # sorted dataframe. Then we need to let i (iterator for a) move to the next value
            
        elif a["Candidate"][i] > b["Candidate"][j]:
            j += 1
            b_unmatched += 1
            # Converse is also true
            
    print("Unmatched in a: ", a_unmatched, "\nUnmatched in b: ", b_unmatched)
    print("Combined dataframe shape: ", combined.shape, "\n")
        
    return combined
```

```python
combined2004 = combine(ls2004sorted, ld2004sorted)
combined2009 = combine(ls2009sorted, ld2009sorted)
combined2014 = combine(ls2014sorted, ld2014sorted)
combined2019 = combine(ls2019sorted, ld2019sorted)
```

```python
# Code for sorting

combined2014sorted = combined2014.sort_values(["State", "Constituency", "Winner"], ascending=[True, True, False])
# Sort states and constituencies in ascending order and sort winner column in descending order 
# Winner is sorted in ascending order to get the name of winner on top with each constituency
```

```python
combined2004.sort_values(["State", "Constituency", "Winner"], ascending=[True, True, False], inplace=True)
combined2009.sort_values(["State", "Constituency", "Winner"], ascending=[True, True, False], inplace=True)
combined2014.sort_values(["State", "Constituency", "Winner"], ascending=[True, True, False], inplace=True)
combined2019.sort_values(["State", "Constituency", "Winner"], ascending=[True, True, False], inplace=True)

combined2004.reset_index(drop=True, inplace=True)
combined2009.reset_index(drop=True, inplace=True)
combined2014.reset_index(drop=True, inplace=True)
combined2019.reset_index(drop=True, inplace=True)
```
